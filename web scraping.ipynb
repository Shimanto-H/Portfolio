```python
from bs4 import BeautifulSoup
```

```python
from urllib.request import urlopen
```

```python
url = "https://en.wikipedia.org/wiki/Main_Page"
```

```python
page = urlopen(url)
```

```python
html=page.read().decode("utf-8")
```

```python
soup= BeautifulSoup(html,"html.parser")
```

```python
print(soup.get_text())
```
